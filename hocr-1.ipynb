```python

# Load the image
image_path = '/kaggle/input/testing-img/iY2Tb.png'
image = cv2.imread(image_path)

# Perform OCR using pytesseract.image_to_data
detection_data = pytesseract.image_to_data(image, output_type=pytesseract.Output.DICT)

# Create a new HTML document
output_html = BeautifulSoup('<html><body></body></html>', 'html.parser')
output_body = output_html.body

# Initialize variables to track the current line and its position
current_line = ''
current_line_top = 0

# Iterate through the detected text and positional information
for i, text in enumerate(detection_data['text']):
    if text.strip():  # Check if the text is not empty or whitespace
        left = detection_data['left'][i]
        top = detection_data['top'][i]
        width = detection_data['width'][i]
        height = detection_data['height'][i]

        # Check if the text is on the same line as the previous text
        if top == current_line_top:
            current_line += ' ' + text
        else:
            # Create a new div element for the current line with position information
            div = output_html.new_tag('div')
            div['style'] = f'position: absolute; left: {left}px; top: {top}px; width: {width}px; height: {height}px;'
            div.string = current_line
            output_body.append(div)

            # Update the current line and its position
            current_line = text
            current_line_top = top

# Create a div element for the last line
div = output_html.new_tag('div')
div['style'] = f'position: absolute; left: {left}px; top: {top}px; width: {width}px; height: {height}px;'
div.string = current_line
output_body.append(div)

# Save the HTML file with preserved structure
output_html_path = '/kaggle/working/output_2.html'
with open(output_html_path, 'w', encoding='utf-8') as html_file:
    html_file.write(str(output_html))


```
